```python
def calculate_monthly_weights(data, dim, target_timemean):
    """
    
    
    """
    
    groupby_string = ".".join([dim, target_timemean])

    month_length         = data[dim].dt.days_in_month
    month_length_grouped = month_length.groupby(groupby_string)
    weights              = month_length_grouped / month_length_grouped.sum()

    return weights
```

```python
def test_calculate_monthly_weights():
    leap_year = 2016
    norm_year = 2015
    
    leap_year_times = pd.date_range(str(leap_year)+"-01-01", str(leap_year)+"-12-31", name="time")
    norm_year_times = pd.date_range(str(norm_year)+"-01-01", str(norm_year)+"-12-31", name="time")
    
    leap_year_da = xr.DataArray(leap_year_times, dims=["time"], coords = {"time":leap_year_times})
    norm_year_da = xr.DataArray(norm_year_times, dims=["time"], coords = {"time":norm_year_times})
    
    leap_year_weights = calculate_monthly_weights(leap_year_da, dim="time", target_timemean ="year")
    norm_year_weights = calculate_monthly_weights(norm_year_da, dim="time", target_timemean ="year")
    
    leap_year_weights_calculated = leap_year_times.daysinmonth/np.sum([monthrange(leap_year, x)[1]**2 for x in range(1,13)])
    norm_year_weights_calculated = norm_year_times.daysinmonth/np.sum([monthrange(norm_year, x)[1]**2 for x in range(1,13)])
    
    assert (leap_year_weights_calculated == leap_year_weights).all() 
    assert (norm_year_weights_calculated == norm_year_weights).all()
    
    
    
```
